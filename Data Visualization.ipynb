```python
#What is Data Visualization?
#Imagine you have a list of numbers or data, like students’ scores, sales figures, or daily temperatures. Instead of just staring at rows and columns of numbers, data visualization helps you turn those numbers into pictures like charts, graphs, or plots. These pictures make it easier to see patterns, trends, and important points at a glance.

#Why is it Important?
#•It helps you spot trends quickly (e.g., sales are increasing each month).

#•You can compare different groups easily (e.g., boys vs. girls performance).
#It makes your presentation look professional and easier to understand.

#Popular Python Libraries for Visualization
#1.Matplotlib
#A flexible tool to make basic charts and graphs, like bar charts, line charts, and pie charts. Think of it like drawing using a ruler and pencil — you control everything.
#Example using Matplotlib

import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [250, 400, 300, 450]

# Plotting
plt.plot(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales (₦)")
plt.show()

#What This Does:
#Draws a line showing how sales changed from January to April.
#The title, xlabel, and ylabel add information to the graph.
```

```python
#2.Seaborn
#This is a fancier, more stylish version built on top of Matplotlib. It automatically makes graphs prettier and easier to interpret.
#Example using Seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Student': ['John', 'Mary', 'Paul', 'Sandra'],
    'Score': [75, 88, 90, 70]
}
df = pd.DataFrame(data)
# Create a bar chart
sns.barplot(x='Student', y='Score', data=df)
plt.title("Student Test Scores")
plt.show()

#What This Does:
#•Plots student names on the x-axis and their scores on the y-axis.
#•Each bar shows how well a student performed.
#•It's visually clean and professional-looking without extra effort.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Student': ['John', 'Mary', 'Paul', 'Sandra'],
    'Score': [75, 88, 90, 70]
}
df = pd.DataFrame(data)

# Create a bar chart
sns.barplot(x='Student', y='Score', data=df)
plt.title("Student Test Scores")
plt.show()
```

```python
#using Matplotlib in Python:
#1. Plot a Simple Chart
#A simple plot of some numbers.
import matplotlib.pyplot as plt

# Sample data
values = [1, 3, 2, 4]
# Plot the data
plt.plot(values)
plt.title("Simple Chart")
plt.show()
```

```python
#2. Plot a Line Chart
#A line chart showing a trend over time.
import matplotlib.pyplot as plt

# Sample data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temperatures = [30, 32, 33, 31, 29]
# Plot line chart
plt.plot(days, temperatures, marker='o', linestyle='-', color='blue')
plt.title("Daily Temperature")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()
```

```python
#3. Plot a Bar Chart
#A bar chart to compare different categories.
import matplotlib.pyplot as plt

# Sample data
students = ['Alice', 'Bob', 'Charlie', 'David']
scores = [85, 90, 78, 92]
# Plot bar chart
plt.bar(students, scores, color='green')
plt.title("Student Scores")
plt.xlabel("Student")
plt.ylabel("Score")
plt.show()
```

```python
#4. Plot a Scatter Chart
#A scatter plot showing relationship between two numeric variables.
import matplotlib.pyplot as plt

# Sample data
hours_studied = [1, 2, 3, 4, 5]
test_scores = [40, 50, 65, 70, 85]
# Plot scatter chart
plt.scatter(hours_studied, test_scores, color='red')
plt.title("Hours Studied vs Test Score")
plt.xlabel("Hours Studied")
plt.ylabel("Test Score")
plt.grid(True)
plt.show()
```
